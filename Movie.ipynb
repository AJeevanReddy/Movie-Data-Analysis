```python
#Droping unnecessary columns 

u1=u.drop(['color','director_facebook_likes','actor_1_facebook_likes','actor_3_facebook_likes','actor_2_name',
'actor_2_facebook_likes',
'cast_total_facebook_likes',
'actor_3_name',
'duration',
'facenumber_in_poster',
'content_rating',
'country',
'movie_imdb_link',
'aspect_ratio',
'plot_keywords'],axis=1)
```

```python
u1.shape
```

```python
# dropping the rows 
df=u1.isnull().sum(axis=0)*100/len(u1)
df1=list(filter(lambda a: df[a]>=5,df.keys()))
u2=u1.dropna(subset=df1,inplace=False)
u2.shape
```

```python
u1.shape
```

```python
# dropping the rows 
df=u2.isnull().sum(axis=1)
df1=list(filter(lambda a: df[a]>=5,df.keys()))
u3=u2.drop(df1,axis=0)
u3.shape
```

```python
#Filling the NaN values in the 'language' column 
u3['language']=u3['language'].replace(np.nan,'English')
```

```python
# number of retained rows 
(u3.shape[0]/5043)*100
```

```python
#Converting the unit of the budget and gross columns from $ to million $.
u3['gross']=u3['gross']*1000000
```

```python
# Finding the movies with highest profit
u3['profit']=(u3['gross']-u3['budget'])
u3.sort_values(by=['profit'],ascending=False).head(10)
```

```python
u3.shape
```

```python
# Droping duplicate values
u3.drop_duplicates(subset='movie_title',keep='last',inplace=True)
u3.sort_values(by=['profit'],ascending=False).head(10)
```

```python
# code for extracting the top 250 movies as per the IMDb score . storing it in a new dataframe 
# and name that dataframe as 'IMDb_Top_250'
IMDb_Top_250=u3.sort_values(by=['imdb_score'] ,ascending=False)
IMDb_Top_250=IMDb_Top_250[IMDb_Top_250['num_voted_users']>25000].head(250)
IMDb_Top_250['Rank']=range(1,251)
IMDb_Top_250=IMDb_Top_250[['Rank','num_voted_users','movie_title','language']]
IMDb_Top_250
```

```python
#Extracting all the movies in the IMDb_Top_250 dataframe which are not in the English language and store them in a new dataframe named Top_Foreign_Lang_Film.
Top_Foreign_Lang_Film =IMDb_Top_250.loc[IMDb_Top_250['language']!='English']
Top_Foreign_Lang_Film.shape
```

```python
#Finding the best directors
#Group the dataframe using the director_name column.
grouped=u3.groupby('director_name')
top10director=grouped['imdb_score'].mean()
top10director=top10director.sort_values(ascending=False).head(10)
top10director
```

```python
#  code for extracting the first two genres of each movie 
new=u3['genres'].str.split("|",n=2,expand=True)
u3['genres_1']=new[0]
u3['genres_2']=new[1]
#u3['genres_2'].fillna(method='ffill',axis=1)
```

```python
#Group the dataframe using genre_1 as the primary column and genre_2 as the secondary column.
movies_by_segment = u3.groupby(['genres_1', 'genres_2']).agg({'gross': np.mean})
```

```python
#Finding the 5 most popular combo of genres by finding the mean of the gross values using the gross column and store them in a new dataframe named PopGenre.
PopGenre = u3.groupby(['genres_1', 'genres_2']).agg({'gross': np.mean}).reset_index().sort_values(['gross'], ascending=False)
PopGenre.head()
```
